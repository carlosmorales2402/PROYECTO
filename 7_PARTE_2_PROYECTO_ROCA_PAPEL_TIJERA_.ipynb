```python
import random

def obtener_eleccion_computadora():
    """
    Genera una elección aleatoria para la computadora
    """
    opciones = ["roca", "papel", "tijera"]
    return random.choice(opciones)

def obtener_eleccion_jugador():
    """
    Obtiene y valida la elección del jugador
    """
    while True:
        eleccion = input("Elige roca, papel o tijera: ").lower()
        if eleccion in ["roca", "papel", "tijera"]:
            return eleccion
        print("Opción no válida. Por favor, intenta de nuevo.")

def determinar_ganador(jugador, computadora):
    """
    Determina el ganador según las reglas del juego
    """
    if jugador == computadora:
        return "¡Empate!"

    reglas = {
        "roca": "tijera",
        "papel": "roca",
        "tijera": "papel"
    }

    if reglas[jugador] == computadora:
        return "¡Ganaste!"
    return "¡La computadora gana!"

def jugar_roca_papel_tijera():
    """
    Función principal del juego
    """
    print("\n=== ROCA, PAPEL O TIJERA ===")
    print("Gana el mejor de 3")

    puntos_jugador = 0
    puntos_computadora = 0

    while True:
        # Obtener elecciones
        eleccion_jugador = obtener_eleccion_jugador()
        if puntos_jugador == 3 or puntos_computadora == 3:
          print("¡¡¡Felicidades Ganador!!!")
          break


        eleccion_computadora = obtener_eleccion_computadora()

        # Mostrar elecciones
        print(f"\nTú elegiste: {eleccion_jugador}")
        print(f"La computadora eligió: {eleccion_computadora}")

        # Determinar y mostrar resultado
        resultado = determinar_ganador(eleccion_jugador, eleccion_computadora)
        print(resultado)

        # Actualizar puntuación
        if resultado == "¡Ganaste!":
            puntos_jugador += 1
        elif resultado == "¡La computadora gana!":
            puntos_computadora += 1

        # Mostrar puntuación actual
        print(f"\nPuntuación:")
        print(f"Tú: {puntos_jugador} | Computadora: {puntos_computadora}")
        print("\n" + "="*30 + "\n")
```

```python
jugar_roca_papel_tijera()
```
